```python
from google.colab import drive
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
drive.mount('/content/gdrive')
dataPath= '/content/gdrive/MyDrive/PersistenceandRetention2024_Appendix.xlsx'
```

```python
stateRetentions = pd.read_excel(dataPath, header = 2, sheet_name='5. State')
stateRetentions=stateRetentions[stateRetentions.Metric=='Retention Rate']
stateRetentions=stateRetentions[stateRetentions['Age at Entry']=='Total']

newData=[]
for x in stateRetentions.filter(['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']).columns:
  for i in range(0,len(stateRetentions)):
    newData.append([x,stateRetentions[x].iloc[i], stateRetentions['State'].iloc[i]])

newD=pd.DataFrame(newData, columns=['Year','Retention Rate','State'])

sns.lineplot(x=newD['Year'], y=newD['Retention Rate'],hue=newD['State'])
plt.title('Retention Rate Over Time vs State')
plt.xlabel('Year')
plt.ylabel('Retention Rate')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
newD.head(20)
```

```python
data = pd.read_excel(dataPath, header = 2, sheet_name='1. Overall')
data=data[data.Metric=='Retention Rate']
data=data[data['Starting Enrollment Intensity']=='Total']
newData=[]
for x in data.filter(['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']).columns:
  for i in range(0,len(data)):
    newData.append([x,data[x].iloc[i], data['Institution Sector at Entry'].iloc[i]])

newD=pd.DataFrame(newData, columns=['Year','Retention Rate','Institution Sector at Entry'])

sns.scatterplot(x=newD['Year'], y=newD['Retention Rate'],hue=newD['Institution Sector at Entry'])
plt.title('Retention Rate Over Time vs Institution Sector at Entry')
plt.xlabel('Year')
plt.ylabel('Retention Rate')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
newD.head(20)
```
